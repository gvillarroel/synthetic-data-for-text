## Categorical
our model takes one-hot encoded versions of categorical features as an input

```python
df = pd.read_csv('../../../datasets/kingcounty/raw/kc_house_data.csv')
category_columns=("condition", "floors", "grade", "view", "waterfront", "zipcode", "bathrooms", "bedrooms",)
continues_columns=('price', 'sqft_living', 'sqft_lot', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long',
       'sqft_living15', 'sqft_lot15',)
```

```python
df.loc[:, category_columns]
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(df.loc[:, category_columns])
display(enc.categories_)
category_data = enc.transform(df.loc[:, category_columns]).toarray()
display(category_data.shape)
```

```python
display(category_data[0])
display(df.loc[:, category_columns].iloc[0])
```

## Numerical Features
normalized numerical features, we use the gaussian quantile transformation from the scikit-learn library 

```python
import numpy as np
import numpy as np
from sklearn.preprocessing import QuantileTransformer
qt = QuantileTransformer(n_quantiles=10, random_state=0)
continues_data = qt.fit_transform(df.loc[:, continues_columns])
```

```python
display(continues_data[0])
display(df.loc[:, continues_columns].iloc[0])
```

```python
x_in = pd.concat([pd.DataFrame(continues_data, columns=qt.get_feature_names_out()), pd.DataFrame(category_data, columns=enc.get_feature_names_out())], axis=1)
```
