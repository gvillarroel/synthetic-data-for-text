```python
import pandas as pd
import numpy as np
df = pd.read_csv('../../../datasets/kingcounty/raw/kc_house_data.csv');
df.sample(3)
```

```python
category_columns=("condition", "floors", "grade", "view", "waterfront", "zipcode", "bathrooms", "bedrooms",)
df_prep = df.loc[:, ['date', 'price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']].assign(target=(np.random.rand(df.shape[0]) > 0.5)*1).astype({k:"category" for k in category_columns})
```

```python
df_prep.loc[:, df_prep.columns != "target"]
df_prep.loc[:, "date"] = (pd.to_datetime(df_prep["date"]) - pd.to_datetime(df.date.min())).dt.days
```

```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE 

sm = SMOTE(random_state=42)
sm.fit(df_prep.loc[:, df_prep.columns != "target"], df_prep.target)
```
